```python
Champions_df = DataFrame(columns = ['Year','Team'])
Champions_df
```

```python
def Champion (element,year):
    index = str(element).find(".html\">")
    index = index + 7
    Team = str(element)[index:str(element).find("</a><")]
    series = Series (data = [year,Team],index = ['Year','Team'])
    return series
    
```

```python
for x in range (1986,2018):
    html = 'https://www.basketball-reference.com/playoffs/NBA_%d.html' %x
    r = urllib.urlopen(html).read()
    soup = BeautifulSoup(r,'lxml')
    list1=[]
    for text in soup.find_all('p'):
        list1.append(text)
    element = list1[2]
    Champions_df = Champions_df.append(Champion(element,x),ignore_index=True)
#Champions_df = Champions_df[0:33]
```
